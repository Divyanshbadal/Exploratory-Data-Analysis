### Changing the Datatype according to the Datafram

```python
def convert_to_datetime(df, columns, date_format='%m/%d/%Y %H:%M'):
    for column in columns:
        df[column] = df[column].str.replace('-', '/')
        df[column] = pd.to_datetime(df[column], format=date_format, errors='coerce')
    return df

# List of columns to convert
columns_to_convert = [
    'order_approved_at',            #1
    'order_delivered_customer_date',#2
    'order_estimated_delivery_date',#3
    ]

# Apply the UDF to your dataframe
comp = convert_to_datetime(comp, columns_to_convert)
```

### 2. Performing Customers/sellers Segmentation

#### a. Divide the customers into groups based on the revenue generated 

```python
comp['customer_segment']=comp['payment_value'].apply(lambda x : 'Very_Low' if x <= 80  else 'Low'
                           if 80 < x <= 160 else 'Moderate'
                           if 160< x <=260  else 'High'
                            if 260 < x <=400 else ' Very_High')
comp[['customer_segment','payment_value']]  
```

#### b. Divide the sellers into groups based on the revenue generated

```python
sell_seg=comp.groupby(['seller_id'])['payment_value'].sum().sort_values(ascending=False).reset_index()
```
